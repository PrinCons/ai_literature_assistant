```python
import requests 
import xml.etree.ElementTree as ET
import datetime
```

```python
url = 'http://export.arxiv.org/api/query?search_query=all:electron&start=0&max_results=1'
url = 'http://export.arxiv.org/api/query?search_query=ti:rocket+OR+abs:rocket&start=0&max_results=10'
url = 'http://export.arxiv.org/api/query?search_query=all:rocket+science+ANDNOT+au:Johnson&start=0&max_results=10'
arxiv_response = requests.get(url)
```

```python
if arxiv_response.status_code == 200:
    print(arxiv_response.text)
```

```python
root = ET.fromstring(arxiv_response.text)

# 
# for entry in root.findall('atom:entry', ns):
#     q = entry.find('atom:title', ns).text.strip()
```

```python
for entry in root:
    print(entry)
```

```python
ns = {'atom': 'http://www.w3.org/2005/Atom'}
for entry in root.findall('atom:entry', ns):
    title = entry.find('atom:title', ns).text.strip()
    summary = entry.find('atom:summary', ns).text.strip()
    authors = [author.find('atom:name', ns).text.strip()
                for author in entry.findall('atom:author', ns)]
    link = entry.find('atom:link', ns).attrib['href']
    id = entry.find('atom:id', ns).text.strip()
    # total_results = entry.find('atom:totalResults', ns).text.strip()
    updated = entry.find('atom:updated', ns).text.strip()
    updated = datetime.datetime.strptime(updated, '%Y-%m-%dT%H:%M:%SZ').date()
    # entry = entry.find('atom:entry', ns)

    print("Title:", title)
    print("Authors:", ', '.join(authors))
    print("Summary:", summary[:20])
    print("Link:", link)
    print("ID:", id)
    print("Updated:", updated)
    # print("entry", entry)
    # print("Total Results:", total_results)
```
